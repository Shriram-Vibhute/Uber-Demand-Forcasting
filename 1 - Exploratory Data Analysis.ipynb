```python
# Importing necessary libraries
import numpy as np
import pandas as pd
import dask.dataframe as dd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# paths for the three dfs
df_jan_path = "yellow_tripdata_2016-01.csv"
df_feb_path = "yellow_tripdata_2016-02.csv"
df_mar_path = "yellow_tripdata_2016-03.csv"

# load the dataframes
df_jan = dd.read_csv(df_jan_path, assume_missing=True)
df_feb = dd.read_csv(df_feb_path, assume_missing=True)
df_mar = dd.read_csv(df_mar_path, assume_missing=True)
```

```python
# visualize the chunks using the task graph
df_jan.visualize()
```

```python
# shape of all the three datasets
print("Shape of january dataset", df_jan.shape)
print("Shape of february dataset", df_feb.shape)
print("Shape of march dataset", df_mar.shape)
```

> *We can concat the three dataframes as the number of columns are same in all the three dfs*

```python
# concat the three dataframes as one
df_final = dd.concat([df_jan, df_feb, df_mar], axis=0).drop(columns=['extra', 'mta_tax', 'tip_amount', 'tolls_amount', 'improvement_surcharge', 'total_amount', 'payment_type'])
```

> *These dropped columns could be sutaible for `Fare prediction` but not for demand forcasting*

```python
df_final["tpep_pickup_datetime"] = dd.to_datetime(df_final["tpep_pickup_datetime"], yearfirst=True, format="mixed")
df_final["tpep_dropoff_datetime"] = dd.to_datetime(df_final["tpep_dropoff_datetime"], yearfirst=True, format="mixed")
```

```python
# create new columns - month num, date of week num, hour
df_final["pickup_months"] = df_final["tpep_pickup_datetime"].dt.month
df_final["pickup_day_of_week"] = df_final["tpep_pickup_datetime"].dt.dayofweek
df_final["pickup_hour"] = df_final["tpep_pickup_datetime"].dt.hour
```

```python
# create new columns - month num, date of week num, hour
df_final["drop_months"] = df_final["tpep_dropoff_datetime"].dt.month
df_final["drop_day_of_week"] = df_final["tpep_dropoff_datetime"].dt.dayofweek
df_final["drop_hour"] = df_final["tpep_dropoff_datetime"].dt.hour
```

```python
# plot the number of pickups
pickups_every_3_days = (
    df_final
    .set_index('tpep_pickup_datetime')
    .resample("3D") # A groupby like method which groups every 3 pickup_datetime
    .size()
    .compute()
)
```

```python
len(pickups_every_3_days)
```

```python
# plot the graph of pickups
sns.lineplot(pickups_every_3_days)
plt.xticks(rotation=45)
plt.show()
```

> *2016-01-22 to 2016-01-25 - In Jan 2016 due to snow blizard the pikups are less in count*

```python
# pickups for each hour of the day
pickups_each_hour = (
    df_final
    .groupby(["pickup_hour","pickup_day_of_week"])["VendorID"]
    .count()
    .compute()
)
```

```python
pickups_each_hour
```

```python
pickups_each_hour = pickups_each_hour.reset_index() # Removing multi-index
pickups_each_hour.rename(columns={"VendorID":"Number of Pickups"},inplace=True)
```

```python
day_names = ["Mon", "Tue", "Wed", "Thu", "Fri", "Sat", "Sun"]
replacement_mapper = {k:v for k,v in enumerate(day_names)}
replacement_mapper
```

```python
pickups_each_hour.replace({"pickup_day_of_week":replacement_mapper},inplace=True)
```

```python
# plot the lineplot
fig = plt.figure(figsize=(12,6))
sns.lineplot(pickups_each_hour, x="pickup_hour", y="Number of Pickups",hue="pickup_day_of_week",hue_order=day_names)
plt.show()
```

> Key Observations on Uber Pickup Patterns:
> Weekday Patterns:
> - Peak hours occur during:
>   - Morning commute (8:00 AM - 10:00 AM)
>   - Evening commute (5:00 PM - 10:00 PM)
> - These spikes align with typical office hours and work commutes
>
> Weekend Patterns:
> - Saturdays show distinct behavior:
>   - Higher activity during late night/early morning (10:00 PM - 12:00 AM)
>   - Lower demand during morning and afternoon hours
>   - Pattern likely reflects leisure and entertainment activities
